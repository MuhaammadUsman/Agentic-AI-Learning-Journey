# **Numpy Indexing, Slicing & Strides**

```python
# import numpy library
import numpy as np
```

```python
# 2-D array of shape (3, 3)\
arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
    
])
```

```python
print(arr)
```

```python
print(arr.ndim) # number of dimensions
```

```python
# indexing of 2-D array
print(arr[0, 1]) # element at row 0 and column 1
```

```python
print(arr[2,2]) # element at row 2 and column 2
```

```python
print(arr[-1, -1]) # element at last row and last column
```

```python
print(arr[-1, -2]) # element at last row and second last column
```

```python
print(arr[-3,-3]) # element at first row and first column
```

```python
# General form of indexing: arr[start : stop : step]
print(arr[0:2, 0:2]) # subarray of first 2 rows and first 2 columns
```

```python
print(arr[0:2, 1:3]) # subarray of first 2 rows and last 2 columns
```

```python
print(arr[::2, ::2]) # subarray of every second row and every second column
```

```python
slice_arr = arr[0, :2] # view not copy
slice_arr[0] = 999
print(arr[0]) # original array is modified due to view
```

```python
copy_arr = arr[1].copy() # copy of second row
copy_arr[0] = 888 # modifying copy does not affect original array
print(copy_arr) # modified copy
```

```python
print(arr)
```
