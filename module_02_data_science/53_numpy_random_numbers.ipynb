# **Numpy Random Numbers, Reproducibility & Performance**
- **Seed:** initial value that helps to determine the sequence of random numbers

```python
import numpy as np
```

```python
# similate 100K coin flips.
rng = np.random.default_rng(seed = 42) # create random number generator with fixed seed for reproducibility
flips = rng.integers(0, 2, size = 1000000) # generate 1 million random integers (0 or 1)
prob_heads = np.mean(flips) # calculate mean which gives probability of heads (1)
print(f"Estimated probability of heads: {prob_heads:.4f}")
```

```python
import time 
# python loop
start = time.time()
squares = [i**2 for i in range(1000000)]
end = time.time()
print(f"Python loop time: {end - start:.4f} seconds")
# numpy vectorized operation
start = time.time()
arr = np.arange(1000000)
squares_np = arr ** 2
end = time.time()
print(f"Numpy vectorized time: {end - start:.4f} seconds")
```
