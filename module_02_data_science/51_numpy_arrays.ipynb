# **Creating Arrays in Numpy**

```python
# import numpy library
import numpy as np
```

```python
# create a 1D array
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr_1d)
# create a 2D array
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr_2d)
# create a 3D array
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Array:\n", arr_3d)
```

```python
# create zeros array
zeros_arr = np.zeros((2,3))
print("Zeros Array:\n", zeros_arr)
```

```python
# create ones array
ones_arr = np.ones((3, 4))
print("Ones Array:\n", ones_arr)
```

```python
# create full array
full_arr = np.full((2, 3), 7) # fill with element 7
print("Full Array:\n", full_arr)
```

```python
# creat range of array with step size
range_arr = np.arange(5, 15, 3)# start at 5, end before 15, step size 3
print("Range Array:", range_arr)
```

```python
# create linearly spaced array
linspace_arr = np.linspace(0, 1, 5) # start at 0, end at 1, total 5 elements
print("Linspace Array:", linspace_arr)
```

```python
# create identity matrix
indentity_arr = np.eye(3) # 3x3 identity matrix
print("Identity Matrix:\n", indentity_arr)
```

```python
# create diagonal array
diag_arr = np.diag([1, 2, 3]) # create diagonal array with elements 1, 2, 3
print("Diagonal Array:\n", diag_arr)
```

```python
# create array with type float
float_arr = np.array([1, 2, 3], dtype=float)
print("Float Array:", float_arr)
```

```python
# array with type int
int_arr = np.array([1.5, 2.5, 3.5], dtype=int)
print("Integer Array:", int_arr)
```
