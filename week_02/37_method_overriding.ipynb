# **Method Overridding**
- **Overriding:** Child replaces parent method

```python
# Parent class
class Animal():  
    def speak(self):
        print("Animal speaks")

# Child class inheriting from Animal
class Dog(Animal):  
    
    # Method overriding (same method name as parent)
    def speak(self):
        print("Wooof! Wooof!")

# Create object of Dog class
dog1 = Dog()

# Call speak() method
dog1.speak()   # Calls Dog's version, not Animal's
```
