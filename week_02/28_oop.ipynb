# **Need for OOP**
- Object Oriented Programming: Understanding and tracking large program codes become difficult

**1. Encapsulation:**
- related data is encapsulated into a single object
- functions related to those variables are grouped together.

**2. Object:**
- combination of related variables and functions.
- stores data (attributes)
- performs actions (methods)

**3. Data Abstraction**
- we donot need to show all our data and functions to outside the world.

**Benefits of Data Attribution:**
- promoted code reusability
- enhances modularity
- improves maintainability
- makes programs closer to real-world thinking.
