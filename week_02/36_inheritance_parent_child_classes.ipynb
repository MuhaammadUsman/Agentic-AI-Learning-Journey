# **Inheritance in Python - Parent & Child Classes**
- **Inheritance:** 
- allows a new class to reuse the properties and methods of an existing class.

```python
# Base class (parent class)
class Animal:
    
    # Method defined in parent class
    def speak(self):
        print("Animal speaks")


# Child class (inherits from Animal)
class Dog(Animal):
    pass   # No new code, just inherits everything from Animal


# Create object of Dog class
dog1 = Dog()

# Call method (inherited from Animal)
dog1.speak()
```

```python
# Child class inheriting from Animal (parent class)
class Dog(Animal):
    
    # New method specific to Dog class
    def bark(self):
        print("Wooof!")


# Create an object of Dog class
childdog = Dog()

# Call inherited method from Animal class
childdog.speak()   # comes from parent class

# Call Dog's own method
childdog.bark()    # defined in child class
```
