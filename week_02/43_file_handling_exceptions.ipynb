# **Handling Exceptions During File Operations**

### **Common Issues in the File Handling**
- 1. file not found
- 2. wrong path
- 3. wrong file mode
- Due to these reasons, program may crash.
### **Exception Handling:** An error that occurs during the execution of a program.

```python
with open ("missing_file.txt", "r") as file:
    print(file.read())
# this will raise a FileNotFoundError since the file does not exist
```

```python
# to handle this exception, we can use a try-except block
try:
    with open ("missing_file.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("The file was not found. Please check the file name and try again.")
```

```python
# file name, name, and age variables for demonstration purposes
filename = "existing_file.txt"
name = "Muhammad Ali"
age = 30
```

```python
# try to read existing file, if missing create it
try:
    with open(filename, "r") as myfile:
        print(myfile.read())

except FileNotFoundError:
    with open(filename, "w") as myfile:
        myfile.write(f"Name: {name}\nAge: {age}")
    with open(filename, "r") as myfile:
        print(myfile.read())
        
```
