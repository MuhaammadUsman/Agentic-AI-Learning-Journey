# **Reading from & Writing to Files in Python**
### **Data Formats:**
- 1. Text Files: textual data
- 2. CSV: comma-separated values. data is stored in tabular form separated by commas.
- 3. JSON: Java-Script Object Notation. used in web applications.

```python
# file opening from current directory
file = open("file.txt", "r")
print(file.read())
```

```python
# closing the file
file.close()
```

```python
# open the file and rewrite it
file = open("file.txt", "w")
file.write("This is a new line.")
file.close()
```

```python
file = open("file.txt", "r")
print(file.read())
```
