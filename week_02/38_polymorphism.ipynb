# **Polymorphism in Python - Overriding & Overloading**
- **Polymorphism:** refers to the ability of function or method to exist in multiple forms.
- Example: Animal (Dog, Cat, Cow, etc)

```python
# Define Dog class
class Dog:
    
    # Method for Dog sound
    def speak(self):
        print("Wooof")

# Define Cat class
class Cat:
    
    # Method for Cat sound
    def speak(self):
        print("Meooow")

# Loop through different objects
for animal in [Dog(), Cat()]:
    
    # Call the same method on different objects
    animal.speak()
```

```python
# Parent class
class Vehicle:
    
    # Method in parent class
    def move(self):
        print("Vehicle is moving")

# Child class inheriting from Vehicle
class Car(Vehicle):
    
    # Overriding parent method
    def move(self):
        print("Car is driving")

# Create object of Car class
car1 = Car()

# Call overridden method from Car class
car1.move()

# Create object of Vehicle class
car2 = Vehicle()

# Call original method from Vehicle class
car2.move()
```

```python
# Define a class named Calculator
class Calculator:
    
    # Method using *numbers
    # *numbers collects multiple arguments into a tuple
    def add(self, *numbers):
        
        # sum() adds all values inside the tuple
        return sum(numbers)

# Create object of Calculator class
calc = Calculator()

# Call method with 1 number
print(calc.add(5))

# Call method with 2 numbers
print(calc.add(5, 10))

# Call method with 3 numbers
print(calc.add(5, 10, 15))
```
