# **WHILE Loop**
- Loop execute as long as specified condition remains true.
- Loop break when condition false
- Loop must be ended

|        FOR Loop                 |            WHILE Loop              |
|---------------------------------|------------------------------------|
| Khnown number of repetitions    | Unkhnown, depends on condition     |
| e.g: Counting to 10     | Waiting for input |

```python
# example
count = 1
while count <=5:
    print(count)
    count += 1
```
