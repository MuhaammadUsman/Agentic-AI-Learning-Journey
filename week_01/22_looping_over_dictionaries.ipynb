# **Looping Over Data Structure-Dictionaries**
- Used to access structure data
- modifying enteries in place
- summerizing or filtering info

```python
# 1. access keys from dict.
student = {"name": "Ali", "age": 20, "grade": "A"}
for key in student:
    print(key)
```

```python
# 2. access values from dict.
student = {"name": "Ali", "age": 20, "grade": "A"}
for key in student:
    print(student[key])
```

```python
# 3. access both keys and values from dict.
student = {"name": "Ali", "age": 20, "grade": "A"}
for key in student:
    print(key, ":", student[key])
```

```python
# filter required data from dict.
student_list = [
    {"name": "Ali", "age": 20, "grade": "A"},
    {"name": "Sara", "age": 22, "grade": "B"},
    {"name": "John", "age": 19, "grade": "A"},
    {"name": "Emily", "age": 21, "grade": "C"},
]

for student in student_list:
    print(f"Name: {student.get('name', '')} | Grade: {student.get('grade', '')}")
```
