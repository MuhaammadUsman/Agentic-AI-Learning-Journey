### **Python Operators: Arithmetic, Comparison, Logical and Simple Arithmetic**
- Two Types: 1. Unary operators 2. Binary Operators

### 1. Mathematical-Arithmetic Operators

```python
print(5 + 3) # addition
print(5 - 3) # subtraction
```

```python
print(2 * 5) # multiply
print( 10 / 2) # divide
print(13 // 2) # floor divide
```

```python
print(10 % 3) # remainder
print(2 ** 3) # exponent
```

### 2. Comparison

```python
print( 5 == 5) # equality
print( 7 != 3) # inequality
print(10 > 5) # greater than
print(15 < 10) # less than
```

### 3. Logical Operators

```python
print(True and False) # AND-both true
print(True or False) # OR-at least one True
print(not True) # NOT-Reverse result
```

# 4. Simple Arithmetic Operators

```python
x = 10
print(x)

x += 5 # same as x = x + 5
print(x)

x -= 3 # same as x = x - 3
print(x)

x *= 2 # same as x = x * 2
print(x)

x /= 4 # same as x = x / 4
print(x)
```
