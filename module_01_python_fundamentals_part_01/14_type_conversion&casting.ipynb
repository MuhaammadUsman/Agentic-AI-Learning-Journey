## **Type Conversion & Casting in Python**

- We can convert one data type into another data type
- int(): convert into integer
- float(): convert into decimal point
- str(): convert into string
- bool(): convert into boolean

```python
age = "30"
type(age)
```

```python
# convert into integer
age = int(age)
type(age)
```

### Input Function ()
- input() take input from user.
- always returns a string type.

```python
# example
age = input("Please enter your age")
print(age)
type(age)
```

```python
# conversion
myAge = int(age)
print(myAge)
type(myAge)
```

### Real-life Example

```python
# Real-life example
name = input("Please write your name here")
age = int(input("Please enter your age here"))
print("Hello", name)
print("Afer 5 years, you will be", age + 5)
```
