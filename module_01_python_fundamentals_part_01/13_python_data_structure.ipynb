### **Topic: Python Data Structure - Lists, Tuples, Sets and Dictionaries**

## **1. Lists []**
- Ordered and mutable
- Allow duplicate values
- Hold multiple values

### 1. Example of lists

```python
fruits = ["apple", "mango", "banana", "apple"]
print(fruits)
```

### 2. Accessing Lists-Indexing

```python
print(fruits[1]) # start from left: +ve indexing
print(fruits[-1]) # start from right: -ve indexing
```

### 3. Lists Methods

```python
# 1. add items
fruits.append("orange") # orange place at last
print(fruits)
```

```python
# 2. Insert at Index
fruits.insert(1, "grapes")
print(fruits)
```

```python
# 3. Removing items
fruits.remove("apple")
print(fruits)
```

```python
# 4. Remove Last item
fruits.pop()
print(fruits)
```

### 4. Looping through List

```python
for fruit in fruits:
    print(fruit)
```

## **2. Tuples()**

- Ordered
- Cannot modify after creation

### 1. Example

```python
colors = ("red", "green", "blue", "white")
print(colors)
```

### 2. Access Tuple

```python
colors[3]
```

```python
# colors[1] = "black" # cannot modify or change, cause error
```

### 3. Tuple Methods

```python
numbers = (1,2,3,4,5,5,6,7,6,5,5)
print(numbers)
```

```python
# count values and position of element 
print(numbers.count(5))
print(numbers.index(2)) # element 2 on position-1
```

## **3. Sets {}**

- Store unique unordered values
- No duplicates
- Fast operation

### 1. Example

```python
my_set = {1,2,3,2,3,4,5,6,5}
print(my_set) # no duplicate
```

### 2. Set Methods

```python
# add new value 
my_set.add(7)
my_set
```

```python
# remove value
my_set.remove(3)
my_set
```

```python
# remove value
my_set.pop()
```

```python
my_set
```

## 3. Set Operations


```python
set_A = {1,2,3}
set_B = {3,4,5}
```

```python
# union-combine all unique values
print(set_A.union(set_B))
```

```python
# intersection-common in both
print(set_A.intersection(set_B))
```

```python
# difference-present in 1st set not in 2nd set
print(set_A.difference(set_B))
```

## **4. Dictionary**

- Key
- Value

### 1. Example

```python
student = {
    "name": "Muhammad Usman",
    "age": 30,
    "Course": "Data Science" 
}
```

```python
print(student)
```

### 2. Access Values

```python
# two way to access values against keys
print(student["name"])
print(student.get("age"))
```

### 3. Dictionary Methods

```python
student["age"] = 25 # update age
student["city"] = "Lahore" # add 
student.pop("Course") # remove value
print(student)
```

### 4. Looping through Dict.

```python
for key, value in student.items():
    print(key, ":", value)
```
