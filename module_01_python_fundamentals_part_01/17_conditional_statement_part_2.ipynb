### Grading

```python
# Grading system using if-elif-else statements
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Try Again!")
```

### **Nested if**
- An if statement placed inside another if statement

```python
age = 17
country = "Pakistan"
if age >= 18:
    if country == "Pakistan": # nested if statement
        print("You can vote in Pakistan.")
```

```python
age = int(input("What is your age?."))
if age >= 18:
    country = input("What is your country?.")
    if country == "Pakistan": # nested if statement
        print("You can vote in Pakistan.")
    else: 
        print("You can not vote in Pakistan.")
else:
    print("Your age is not enough for vote.")
```

```python
# combine conditions using logical operators
age  = int(input("What is your age?"))
country = input("What is your country?").capitalize() # capitalize first letter

if age >= 18 and country == "Pakistan": # combine conditions using logical operator 'and'
    print ("Eligible voter")
else:
    print("Not eligible voter")
```

```python
# multiple conditions using logical operators
marks = int(input("Enter your marks:"))
if marks >= 90:
    print("Grade: A+")
elif marks >= 80:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
else:
    print("F")
```
