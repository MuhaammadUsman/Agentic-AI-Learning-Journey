# **Local vs Global Variables**
- 1. Local Variable: declared inside a function
- 2. Global Variable: declared outside the function (in the main program)

```python
# local variable
def show():
    x = 5 # local
    print(x)
show()
```

```python
# global variable
x = 10 # global
show() # call local variable
print(x) # call global variable
```
