# **Functions**
- reusable block of code that perform a specific task.
- Benefits: 1. increase reuse-ability of code. 2. Helps make programs organized
- Calling a Function: After defining a function, we call it to execute.

```python
# define a func
def greet():
    print("Hello, World!")
# call the func
greet()
```

```python
# def a func with value
def greet(name):
    print("Hello,:", name)
greet("Usman")
```

```python
greet("Azfar")
```
