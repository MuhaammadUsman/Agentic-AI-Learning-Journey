# **Encapsulation-Public & Private Attributes**
- encapsulation means keeping all the data attributes and methods of an object inside that object.
- **Private Attributes:**
- those attributes that can only be used within the class's own method.
- not used in global scale.

```python
# Accessing private attribute outside the class
class Student:
    def __init__(self, name, grade):
        self.__name = name  # Private attribute
        self.__grade = grade # Private attribute
# object of the Student class
student1 = Student("Usman", "A+")
print(student1.__name)  # This will raise an AttributeError
```

```python
# Accessing private atrribute using method inside the class
class Student:
    def __init__(self, name, grade):
        self.__name = name  # Private attribute
        self.__grade = grade # Private attribute
    # Method to access private attribute
    def get_name(self):
        return self.__name
# object of the Student class
student1 = Student("Nida", "A+")
print(student1.get_name())  # This will print "Usman"
```
