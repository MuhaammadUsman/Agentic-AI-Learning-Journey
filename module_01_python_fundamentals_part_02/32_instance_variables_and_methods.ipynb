# **Instance Variable & Instance Method**
- **Instance:**
- refers to an object created from a class.
- each object has its own attributes and methods

```python
# Define a class named Student
class Student:
    
    # Constructor method (runs automatically when object is created)
    def __init__(self, name, age, grade):
        self.name = name    # store student's name
        self.age = age      # store student's age
        self.grade = grade  # store student's grade
    
    # Method to display student information
    def display_info(self):
        print(f"Name: {self.name}")   # print name
        print(f"Age: {self.age}")     # print age
        print(f"Grade: {self.grade}") # print grade
    
    # Method to check exam eligibility
    def is_eligible(self):
        if self.age >= 15:   # condition for eligibility
            print(f"{self.name} is eligible for the exam.")
        else:
            print(f"{self.name} is not eligible for the exam.")
```

```python
# Create student objects with name, age, and grade
s1 = Student("Ali", 16, "A")
s2 = Student("Ahmed", 13, "B")

# Display information and check eligibility for both students
s1.display_info()
s1.is_eligible()

s2.display_info()
s2.is_eligible()
```
