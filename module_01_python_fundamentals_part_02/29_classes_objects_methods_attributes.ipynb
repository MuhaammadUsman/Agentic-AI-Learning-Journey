# **Core Concepts-Classes, Objects, Methods & Attributes**

### **1. Classes:**
- is a blueprint for creating objects. this is basically template or layout or design.
- define attributes (data) & methods (behaviours)
### **2. Object:**
- is a real-instance of a class.
### **3. Attributes:**
- data stored in the object (name, age)
### **4. Methods:**
- actions that object can perform (speak(), run()).
- together they define what the class is and what it does.

```python
# example
class Car:
    def __init__ (self, color):
        self.color = color
    def drive(self):
        print(f"The {self.color} car is driving.")

# first object
my_car = Car("red")
my_car.drive()  # Output: The red car is driving.
# second object
my_brother_car = Car("blue")
my_brother_car.drive()  # Output: The blue car is driving.
```
