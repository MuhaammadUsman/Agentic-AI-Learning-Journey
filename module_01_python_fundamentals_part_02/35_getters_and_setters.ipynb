# **Using Getters and Setters in Python**

### **Getters:**
- are functions used to retrieve the data attributes of a class.
### **Setters:**
- are functions used to assign or modify the data attributes of a class.

```python
# Define a class named Student
class Student:
    
    # Property decorator (allows accessing method like an attribute)
    @property
    def grade(self):
        return self.__grade   # return private grade value

    # Constructor (runs when object is created)
    def __init__(self, name, grade):
        self.__name = name    # Private attribute (name)
        self.__grade = grade  # Private attribute (grade)

    # Getter method for name
    def get_name(self):
        return self.__name

    # Getter method for grade
    def get_grade(self):
        return self.__grade

    # Setter method for grade with validation
    def set_grade(self, grade):
        # Check if grade is between 0 and 100
        if 0 <= grade <= 100:
            self.__grade = grade   # update grade
        else:
            print("Invalid grade. Please enter a value between 0 and 100.")
```

```python
# Create a Student object with initial grade = 85
student = Student("Ali", 85)

# Update grade to 18 using setter method
student.set_grade(120)

# Print updated grade
print(student.get_grade())  # Output 
```
