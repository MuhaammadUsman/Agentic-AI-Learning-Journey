# **Importing & Using Python Modules**
- **Module** is a file containing large python code (functions, variables)
- We can import and reuse them.
- **Common Built-in Modules**
- 1. math - import math
- 2. random - import random
- 3. os - operating system interaction

```python
# math module
import math
print(math.sqrt(30))
```

```python
# import just functions from math module
from math import sqrt
print(sqrt(25))
```

```python
# custom module
import my_module
print(my_module.greet(("Nida")))
```
