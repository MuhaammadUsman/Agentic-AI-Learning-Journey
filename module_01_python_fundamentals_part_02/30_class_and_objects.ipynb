# **Defining Class & Creating Objects**

```python
# create class and object
class Car:
    def drive(self): # method
        print("Car is moving")
```

```python
car1 = Car() # create object
car1.drive()
```

```python
# with attributes
class Car:
    color = "White"
    def drive(self):
        print(f"{self.color} car is moving")
car1 = Car()
car1.drive()
```

```python
# Define a class named Car
class Car:
    # Class attribute (default value shared by all objects unless changed)
    color = "red"
    # Method to display car movement
    def drive(self):
        # self refers to the current object
        # self.color gets the color of that specific object
        
        print(f"{self.color} car is moving") 
        
    # Method to change the color of the car
    def setColor(self, new_color):
        # Assign new color to the object (instance attribute)
        self.color = new_color
        
# Create an object (instance) of Car class
car2 = Car()
# Call method to change color from default "red" to "Black"
car2.setColor("Black")
# Call method to print car movement with updated color
car2.drive()
```
