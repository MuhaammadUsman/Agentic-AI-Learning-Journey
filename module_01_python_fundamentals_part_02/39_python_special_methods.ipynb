# **Python Special Methods**
- **Special Functions:**
- Built-in behaviour automatically associated with certain operations.

```python
# Define a class named Book
class Book:
    
    # Constructor method
    # Runs automatically when object is created
    def __init__(self, title, author, year):
        
        # Store book information in object attributes
        self.title = title
        self.author = author
        self.year = year

    # Special method (__str__)
    # Controls how object is displayed when printed
    def __str__(self):
        
        # Return formatted string representation of object
        return f"{self.title} by {self.author} ({self.year})"

# Create object of Book class
book1 = Book(
    "Hundred Years of Solitude",
    "Gabriel Garcia Marquez",
    1967
)

# Print object
print(book1)
```

```python
# Define Playlist class
class Playlist:
    
    # Constructor method
    def __init__(self, songs):
        
        # Store list of songs
        self.songs = songs
    
    # Special method for len()
    def __len__(self):
        
        # Return number of songs
        return len(self.songs)
    
    # Special method for + operator
    def __add__(self, other):
        
        # Combine songs from two playlists
        return Playlist(self.songs + other.songs)

# Create first playlist object
p1 = Playlist(["S1", "S2"])

# Create second playlist object
p2 = Playlist(["S3"])

# Print number of songs in p1
print(len(p1))

# Combine playlists and print total songs
print(len(p1 + p2))

# Print number of songs in p2
print(len(p2))
```
