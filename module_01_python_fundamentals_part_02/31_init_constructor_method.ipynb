# **_'_init__() Constructor Method**

```python
# Define a class named Dog
class Dog:
    def __init__(self): # Constructor method (runs automatically when object is created)
        # self refers to the current object
        self.name = "Buddy" # Initialize (assign) a default name to the dog
    # Method to make the dog bark
    def bark(self):
        print(f"{self.name} says Woof!")  # Access the object's name using self.name

dog1 = Dog() # Create an object (instance) of Dog class
dog1.bark() # Call the bark method
    
```

```python
# Define a class named Car
class Car:
    
    # Constructor method (runs automatically when an object is created)
    def __init__(self, brand, color):
        # self refers to the current object
        
        # Assign the passed values to instance variables
        self.brand = brand   # store car brand
        self.color = color   # store car color


# Create an object of Car class and pass values
car1 = Car("Toyota", "White")

# Access and print object attributes
print(car1.brand)   # prints brand of car1
print(car1.color)   # prints color of car1
```

```python
# Change the color of an existing object (car1)
car1.color = "Blue"   # updating instance variable of car1

# Create a new object with different values
car2 = Car("Honda", "Red")

# Print values
print("Car1 color:", car1.color)   # updated value
print("Car2 color:", car2.color)   # original value from constructor
print("Car2 brand:", car2.brand)   # brand of car2
```
