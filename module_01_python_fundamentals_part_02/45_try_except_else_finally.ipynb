# **Using try, except, else & finally Blocks**

```python
# Start try block
# Code inside try may generate errors

try: # Take input from user and convert it into integer
    num = int(input("Enter a number: ")) # Divide 10 by entered number
    
    # This can generate ZeroDivisionError if num = 0
    print(10 / num)
    
# Handles error when user enters 0
except ZeroDivisionError:
    print("You cannot divide by zero!")

# Handles error when user enters text instead of number
except ValueError:
    print("That's not a valid number!")

# Handles any other unexpected errors
except Exception as e:
    print(f"An unexpected error occurred: {e}") # Print actual error message
```

```python
try: # Test risky code
    print("Running some code...")
except: # Handle errors
    print("An error occurred.")
else: # Run if no error
    print("No errors occurred.")
finally: # Always execute
    print("This will always run, regardless of errors.")
    
```
