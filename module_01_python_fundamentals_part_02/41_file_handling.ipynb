# **Modes of File Handling - Read, Write & Append**
- **Use of Append:** add new content to the end of a file without deleting existing data.

```python
# open file and read this file
file = open("file.txt", "r")
print(file.read())
```

```python
file.close() # close file after reading
```

```python
# add more text to file using append mode
file = open("file.txt", "a")
file.write("\nThis another line")
file.close()
```
