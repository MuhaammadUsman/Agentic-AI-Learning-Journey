# **Accessing Object Attributes & Methods**

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name # instance variable for the student's name
        self.age = age   # instance variable for the student's age
        self.grade = grade
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")
    
    def is_eligible(self):
        if self.age >= 15 and self.grade >= 60:
            print(f"{self.name} is eligible for the program.")
        else:
            print(f"{self.name} is not eligible for the program.")
```

```python
# creating objects
student1 = Student("Usman", 16, 85)
student2 = Student("Nida", 14, 75)
```

```python
# Acessing attributes
print(student1.name)
print(student2.age)
```

```python
# Calling methods
student1.display_info()
student1.is_eligible()

student2.display_info()
student2.is_eligible()
```
